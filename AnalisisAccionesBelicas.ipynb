```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import sodapy
```

```python
def read_soda(url, numReg):
  lmt = f'?$limit={numReg}'
  if url.split('.')[-1] == 'json':
    df = pd.read_json(url+lmt)
  elif url.split('.')[-1] == 'csv':
    df = pd.read_csv(url+lmt)
  return df
```

```python
eventos = read_soda('https://www.datos.gov.co/resource/39qq-a72j.json', 37147)
```

```python
agg_anio_departamento = eventos.groupby(by=['a_o','departamento'], as_index=False).agg({'id_caso':'count'}).sort_values(by='id_caso', ascending=False)
```

```python
agg_anio_departamento.rename(columns={'id_caso':'eventos', 'a_o':'año', 'departamento':'departamento'}, inplace=True)
```

```python
agg_anio_departamento.reset_index(drop=True, inplace=True)
```

Para análsis se filtra la información de Antioquia para el año 2003

```python
antioquia_2003 = eventos[(eventos['departamento']=='ANTIOQUIA') & (eventos['a_o']==2003)]
```

```python
antioquia_2003
```

```python
antioquia_2003.reset_index(drop=True, inplace=True)
```

```python
antioquia_2003
```

```python
import folium

mapa = folium.Map(location=[6.9638723516, -75.4188331376], zoom_start=8)

for index, row in antioquia_2003.iterrows():
    lat = row['latitud_longitud']['coordinates'][1]
    lng = row['latitud_longitud']['coordinates'][0]
    folium.Marker([lat,lng], popup=row['modalidad']).add_to(mapa)

mapa
```

## Conclusiones:

- La limpieza de datos es crucial para el análisis y visualización de datos de acciones bélicas en el contexto del conflicto armado en Colombia. Los datos limpios aseguran que los análisis sean precisos y las visualizaciones representen fielmente la realidad. Al respecto se eviendio que los datos  presentan un buen estado de calidad en terminos de consistencia, completitud y revelancia. Por lo tanto no fue necesaria la aplicación de acciones de remediación de errores para el análsis realizado.

- El filtrado de datos es fue etapa fundamental en el análisis y visualización de los datos de hechos asociados a acciones bélicas, especialmente cuando se enfoca en un contexto específico como el departamento de Antioquia en el año 2003, un período con alta incidencia de eventos.

- El filtrado de datos permitó enfocar el análisis en los eventos más relevantes, mejorando la precisión y utilidad de los resultados. En el caso del departamento de Antioquia en 2003, el filtrado ayudó a identificar patrones especificos que se corresponden con la Seguridad Democratica del primer periodo presidencial de Alvaro Uribe Velez.

- La georreferenciación es una herramienta esencial para el análisis y visualización de datos de conflictos armados, especialmente en contextos específicos como el espacio y tiempo elegidó, que tuvo una alta incidencia de eventos.
